```python
import os
import json
```

```python
dir1 = r"C:\Users\alyey\Git\BeyondLabels\output\runs\pathvul_run_logs"
unq = set()
for filename in os.listdir(dir1):
    if "except" in filename: continue
    with open(os.path.join(dir1, filename), "r") as f:
        data = json.load(f)
    unq.add(data["prompt_name"])
    if data["prompt_name"] == "prompt_1":
        data["prompt_name"] = "llmql"
    elif data["prompt_name"] == "prompt_2":
        data["prompt_name"] = "baseline"
    with open(os.path.join(dir1, filename), "w") as f:
        json.dump(data, f, indent=4)
```
